# Pandas 보충강의 - 정규표현식으로 문자열 다루기

```python
import pandas as pd
```

```python
movies = pd.read_csv('/content/drive/MyDrive/추천 시스템 입문(inflearn)/data/ml-latest-small/movies.csv', index_col='movieId')
```

```python
movies.sample()
```

```python
# 정규 표현식 이용해서 문자 가져오기 str.extract('(정규표현식)')
movies['year'] = movies['title'].str.extract('(\(\d\d\d\d\))')
```

```python
#괄호 제거
movies['year'] = movies['year'].str.extract('(\d\d\d\d)')
movies
```
